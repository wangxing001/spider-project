```python
import requests
import json
import pprint
from urllib.parse import urlencode
```

```python
def get_content_url(page_num,video_id,sort_type):
    data = {
        "jsonp": "jsonp",
        "pn": page_num,
        "type": "1",
        "oid": video_id,
        "sort": sort_type
    }
    url = "https://api.bilibili.com/x/v2/reply?"+ urlencode(data)
    return url


def get_content(url):
    response = requests.get(url)
    response_json = json.loads(response.text)
    content_list = response_json["data"]["replies"]
    con_list = []
    for item in content_list:
        print(item["member"]['mid'],item["member"]['uname'],item["content"]["message"])
        li = "用户ID:"+ item["member"]['mid'] + "用户名："+ item["member"]['uname'] + "评论内容：" + item["content"]["message"]
        con_list.append(li)
        double_deck = item["replies"]
        if double_deck:
            for item1 in double_deck:
                print(item1["member"]['mid'],item1["member"]['uname'],item1["content"]["message"])
                li1 = "用户ID:"+ item1["member"]['mid'] + "用户名："+item1["member"]['uname'] + "评论内容：" + item1["content"]["message"]
                con_list.append(li1)
    return con_list   


def save_to_file(content):
    with open('content.txt','a',encoding="utf8") as f:
        f.write(content+'\n')
        
        f.close()
     
    
if __name__ == "__main__":
#     eg: https://www.bilibili.com/video/av48762991
    video_id = "48762991"
    for pn in range(108):
        url = get_content_url(pn+1,video_id,0)
        content = get_content(url)
        for item in content:
            save_to_file(item)
```
