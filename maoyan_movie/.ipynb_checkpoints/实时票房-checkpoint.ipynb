```python
import requests
import json

def get_url():
    url = 'https://piaofang.maoyan.com/dashboard'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36"
    }
    response = requests.get(url,headers=headers).text
    print(response)

def get_context(response):
    data = json.loads(response) 
    
    data = data.get('data')
    info_list = []
    for list in data.get('list'): 
#         print(list)        
        movie_name = list.get('movieName') 
        release_info = list.get('releaseInfo')
        seat_rate = list.get('seatRate')
        sumbox_info = list.get('sumBoxInfo') 
        yield (movie_name,release_info,seat_rate,sumbox_info)

if __name__ == '__main__':
    get_url()
    
    url = 'https://box.maoyan.com/promovie/api/box/second.json'
    headers ={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36"
    }
    response = requests.get(url,headers=headers).text
    info_dict = get_context(response)
    for info in info_dict:
        print(info)
```
